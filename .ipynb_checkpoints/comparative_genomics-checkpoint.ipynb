```python
%matplotlib inline
from os import listdir
from os.path import isfile, join
import numpy as np
from glob import glob
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib
import statsmodels.stats.multitest as smt
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
from adjustText import adjust_text
from scipy import stats
import itertools
import matplotlib.patches as mpatches
```

```python
dir_path = '/Users/osipova/Documents/LabDocs/Brood_parasites_analysis/Comparative_BP/'
file_name = '19_assemblies.toga.galGal6.taeGut5.tsv'
file_data = glob(dir_path + file_name)[0]
df = pd.read_csv(file_data, sep='\t')

df['total'] = df['intact'] + df['mutations'] + df['missing']
df['I+UL']  = df['intact'] + df['mutations']

fig, axes = plt.subplots(1, 2, figsize = (11, 5.5))
references = ['galGal6', 'taeGut5']

for i in range(2):
    ref = references[i]
    axi = axes[i]
    dfi = df[df['reference'] == ref]
    sns.barplot(ax=axi, data=dfi, x='total', y='query', color='grey')
    sns.barplot(ax=axi, data=dfi, x='I+UL',  y='query', color='orange')
    sns.barplot(ax=axi, data=dfi, x='intact', y='query', color='blue')
    axi.set_ylabel('')
    axi.set_xlabel('number of intact genes')
    axi.set_title('reference: {}'.format(ref))
    
sns.despine(fig=None, ax=None, left=False, bottom=False, offset=None, trim=False)
fig.tight_layout()

pdf_name = 'new.pdf'
# fig.savefig(pdf_name, bbox_inches='tight')
```

```python
fig, axes = plt.subplots(figsize = (10, 5))

p = ['#67a9cf', '#ef8a62']
sns.set_palette(p)
sns.barplot(data=df, y='intact', x='query', hue='reference')
plt.xticks(rotation=45)
axes.set_xlabel('')
axes.set_ylabel('number of intact genes')

sns.despine(fig=None, ax=None, left=False, bottom=False, offset=None, trim=False)
fig.tight_layout()

pdf_name = 'intact_genes_by_reference.toga.pdf'
# fig.savefig(pdf_name, bbox_inches='tight')
```

```python
dir_path = '/Users/osipova/Documents/LabDocs/Brood_parasites_analysis/Comparative_BP/'
file_name = 'commonNames.gene_stats.togas.taeGut5.tsv'
file_data = glob(dir_path + file_name)[0]
df = pd.read_csv(file_data, sep='\t')

df['total'] = df['intact'] + df['mutations'] + df['missing']
df['I+UL']  = df['intact'] + df['mutations']

fig, axes = plt.subplots(1, 3, figsize = (11, 11))

## contig N50
axi = axes[0]
sns.barplot(ax=axi, data=df, x='contigN50', y='query', color='grey')
axi.set(xscale="log")
axi.set_ylabel('')

## scaffold N50
axi = axes[1]
sns.barplot(ax=axi, data=df, x='scaffoldN50', y='query', color='grey')
axi.set(xscale="log")
axi.set_yticklabels('')
axi.set_ylabel('')

axi = axes[2]
sns.barplot(ax=axi, data=df, x='total', y='query', color='grey')
sns.barplot(ax=axi, data=df, x='I+UL',  y='query', color='orange')
sns.barplot(ax=axi, data=df, x='intact', y='query', color='blue')
axi.set_ylabel('')
axi.set_yticklabels('')
axi.set_xlabel('number of intact genes')
    
sns.despine(fig=None, ax=None, left=False, bottom=False, offset=None, trim=False)
fig.tight_layout()

pdf_name = 'assembly_stats.togas_taeGut5.pdf'
# fig.savefig(pdf_name, bbox_inches='tight')
```
